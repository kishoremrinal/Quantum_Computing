```python
print("MCQ 1")
print("If ψ1 = 1 and ψ2 = -1, what is P = |ψ1 + ψ2|^2 ?")
print("A) 0   B) 1   C) 2   D) 4")

# Calculate for MCQ 1
psi1_mcq1 = 1
psi2_mcq1 = -1
P_mcq1 = abs(psi1_mcq1 + psi2_mcq1)**2

print(f"\nCalculation: P = |{psi1_mcq1} + {psi2_mcq1}|^2 = |{psi1_mcq1 + psi2_mcq1}|^2 = {P_mcq1}")
print(f"The answer for MCQ 1 is: {P_mcq1}")
# answer: A
```

```python
print("\nMCQ 2")
print("If ψ1 = 1 and ψ2 = i, what is P = |ψ1 + ψ2|^2 ?")
print("A) 0   B) 1   C) 2   D) 4")

# MCQ 2 Calculation
psi1_mcq2 = 1
psi2_mcq2 = 1j
P_mcq2 = abs(psi1_mcq2 + psi2_mcq2)**2

print(f"The value of P = |ψ1 + ψ2|^2 is {P_mcq2}. Hence answer is C.")
```

```python
import cmath

print("\nMCQ 3")
print("If ψ1 = 1 and ψ2 = e^{iπ/3}, what is P = |ψ1 + ψ2|^2 ?")
print("A) 1   B) 2   C) 3   D) 4")

# MCQ 3 Calculation
psi1_mcq3 = 1
psi2_mcq3 = cmath.exp(1j * cmath.pi / 3)
P_mcq3 = abs(psi1_mcq3 + psi2_mcq3)**2

print(f"The value of P = |ψ1 + ψ2|^2 is {P_mcq3:.2f}. Hence, answer is C.")
```

```python
print("\nMCQ 4 (Decoherence)")
print("If |ψ1|^2 = |ψ2|^2 = 1 and γ = 0, what is P?")
print("A) 0   B) 1   C) 2   D) 4")

# MCQ 4 Calculation
# When gamma = 0, the interference term vanishes, so P = |psi1|^2 + |psi2|^2
abs_psi1_sq = 1
abs_psi2_sq = 1
gamma_mcq4 = 0

P_mcq4 = abs_psi1_sq + abs_psi2_sq + gamma_mcq4 * 0

print(f"The value of P is {P_mcq4}. Hence, answer is C.")
```

```python
import numpy as np

print("\nMCQ 5")
print("What is P(+x | +z)?")
print("A) 0   B) 1/4   C) 1/2   D) 1")

# Define Z basis
up_z_mcq5 = np.array([1, 0], dtype=complex)

# Define X basis
up_x_mcq5 = (1/np.sqrt(2)) * np.array([1, 1], dtype=complex)

def probability_mcq5(state, outcome):
    return np.abs(np.vdot(outcome, state))**2

# MCQ 5 Calculation
P_mcq5 = probability_mcq5(up_z_mcq5, up_x_mcq5)

print(f"The value of P(+x | +z) is {P_mcq5:.1f}. Hence, option C.")
```

```python
import numpy as np

print("\nMCQ 6")
print("Start in |+z>. Measure X. Get +x. Then measure Z.")
print("What is P(+z)?")
print("A) 0   B) 1/4   C) 1/2   D) 1")

# Define Z basis
up_z_mcq6 = np.array([1, 0], dtype=complex)
down_z_mcq6 = np.array([0, 1], dtype=complex)

# Define X basis
up_x_mcq6 = (1/np.sqrt(2)) * np.array([1, 1], dtype=complex)

def probability_mcq6(state, outcome):
    return np.abs(np.vdot(outcome, state))**2

# Initial state: |+z>
# First measurement: Measure X, and the result is +x. The state collapses to |+x>.
collapsed_state_mcq6 = up_x_mcq6

# Second measurement: Measure Z on the collapsed state. What is P(+z)?
P_mcq6 = probability_mcq6(collapsed_state_mcq6, up_z_mcq6)

print(f"The value of P(+z) after the sequence of measurements is {P_mcq6:.1f}. This corresponds to option C.")
```
