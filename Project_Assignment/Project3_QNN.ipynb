### QNN Problem: Expectation Value Calculation

This notebook will address the problem of finding the expectation value of `(Z @ Z)` for a two-qubit QNN circuit. We'll use the Qiskit library to construct and simulate the quantum circuit.

**Assumptions about the Circuit:**
Given the input state `|00⟩` and the need to express the expectation value in terms of `θ1, θ2, θ3`, and considering the OCR output for the gate types (`Rx`, `Ry`), I will assume the following circuit structure, which is a common ansatz in variational quantum algorithms:

1.  **Qubit 0:** `Rx(θ1)`
2.  **Qubit 1:** `Ry(θ2)`
3.  **Entanglement:** `CNOT` gate with control qubit 0 and target qubit 1.
4.  **Qubit 0:** `Ry(θ3)`

This structure ensures all three parameters are used and provides a typical QNN layout with both single-qubit rotations and entanglement.

### Define and Construct the Quantum Circuit

Now, we'll define the quantum circuit using Qiskit. We'll use symbolic parameters for `θ1`, `θ2`, and `θ3` to derive a general expression for the expectation value. The input state is `|00⟩`.

```python
from qiskit import QuantumCircuit, transpile
from qiskit.quantum_info import Statevector, Operator
from qiskit.circuit import Parameter
import numpy as np

# Define symbolic parameters
theta1 = Parameter('theta1')
theta2 = Parameter('theta2')
theta3 = Parameter('theta3')

# Create a two-qubit quantum circuit (removed classical bits)
qc = QuantumCircuit(2) # 2 qubits

# Apply the gates as per our assumed structure
# Input state is |00>, so no initial gates needed to set it.
qc.rx(theta1, 0)  # Rx(theta1) on qubit 0
qc.ry(theta2, 1)  # Ry(theta2) on qubit 1
qc.cx(0, 1)       # CNOT with control 0, target 1
qc.ry(theta3, 0)  # Ry(theta3) on qubit 0

# Display the circuit
print("Quantum Circuit:")
print(qc.draw(output='text'))
```

### Calculate the Expectation Value of (Z @ Z)

We will calculate the expectation value of `Z ⊗ Z` for the circuit's final state `|ψ⟩`.

```python
from qiskit.quantum_info import SparsePauliOp, Statevector

# Define the (Z @ Z) operator
zz_operator = SparsePauliOp('ZZ')

# Bind parameters to arbitrary values for demonstration
bound_circuit = qc.assign_parameters({theta1: np.pi/2, theta2: np.pi/3, theta3: np.pi/4})

# Get the state vector directly from the circuit
output_statevector = Statevector.from_circuit(bound_circuit)

print("\nOutput Statevector (for example parameters π/2, π/3, π/4):")
print(output_statevector)

# Calculate the expectation value for these sample parameters
expectation_value_sample = output_statevector.expectation_value(zz_operator)

print(f"\nExpectation Value of (Z @ Z) for sample parameters: {expectation_value_sample.real}")
```

### Summary of the Expectation Value

As shown in the code above, calculating the *symbolic* expectation value for a circuit with `Parameter` objects directly within Qiskit can be complex. Typically, it involves manual symbolic manipulation of the matrix representations of the gates and the final state vector.

For a general two-qubit state `|ψ⟩ = α|00⟩ + β|01⟩ + γ|10⟩ + δ|11⟩`, the expectation value of `Z ⊗ Z` is:

`E = <ψ| (Z ⊗ Z) |ψ> = |α|^2 - |β|^2 - |γ|^2 + |δ|^2`

By following the manual symbolic calculation steps outlined in the previous cell, and substituting the complex amplitudes (`α, β, γ, δ`) that would be derived as functions of `θ1, θ2, θ3`, you would arrive at the final symbolic expression for `(Z @ Z)`.
